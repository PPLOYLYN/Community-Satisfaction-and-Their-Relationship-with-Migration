```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

#ดาวน์โหลดฟอนต์
font_path = "Sarabun-Regular.ttf"
fm.fontManager.addfont(font_path)
font_prop = fm.FontProperties(fname=font_path)
plt.rcParams['font.family'] = font_prop.get_name()
```

**กราฟที่ 1:** ระดับความพึงพอใจในแต่ละด้าน (ทั่วประเทศ)

```python
df_filtered = df_survey[
    (df_survey['region'] == 'ทั่วประเทศ') &
    (df_survey['Satisfaction'] != 'รวม')
]
#Create bar chart
plt.figure(figsize=(15, 15))
ax = sns.barplot(
    data=df_filtered,
    x='Percentage_people_satisfied',
    y='Living_condition',
    hue='Satisfaction',
    dodge=True,
    palette='muted'
)

#Add label
for container in ax.containers:
    ax.bar_label(
        container,
        fmt='%.1f',
        padding=10,
        color='black',
        fontsize=8,
        weight='bold'
    )

#Adjust graph
plt.title('ระดับความพึงพอใจในแต่ละด้าน (ทั่วประเทศ)', fontsize=18)
plt.xlabel('ร้อยละของผู้พึงพอใจ', fontsize=14)
plt.ylabel('ด้านของความเป็นอยู่', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='ระดับความพึงพอใจ', fontsize=10, title_fontsize=12)
plt.tight_layout()


plt.savefig("G1.png", dpi=600, bbox_inches='tight')
plt.show()

#Save graph
#files.download("G1.png")
```

**กราฟที่ 2:** ร้อยละระดับความไม่พึงพอใจต่อการเข้าถึงแหล่งเงินทุนดอกเบี้ยต่ำ (ปี2565)

```python
#Filter data for analysis
df_filtered = df_survey[
    (df_survey['Living_condition'] == 'การเข้าถึงแหล่งเงินทุนดอกเบี้ยต่ำ') &
    (df_survey['Satisfaction'] == 'ไม่พึงพอใจเลย')
]

#Create bar chart
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    data=df_filtered,
    x='region',
    y='Percentage_people_satisfied',
    hue='region',
    palette='muted'
)

#Add label
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', label_type='edge', padding=3)

#Adjust graph
plt.title('ร้อยละระดับความไม่พึงพอใจต่อการเข้าถึงแหล่งเงินทุนดอกเบี้ยต่ำ (ปี2565)', fontsize=16)
plt.xlabel('ภูมิภาค', fontsize=14)
plt.ylabel('ร้อยละของผู้พึงพอใจ', fontsize=14)
plt.xticks(rotation=15)
plt.tight_layout()


plt.savefig("G2.png", dpi=600, bbox_inches='tight')
plt.show()

#Save graph
#files.download("G2.png")
```

**กราฟ 3:** ร้อยละความต้องการของประชาชนที่ต้องการให้ภาครัฐเข้ามาช่วยเหลือ (ปี2566) ทั่วประเทศ

```python
#Filter data for analysis
df_filtered = df_popneed[df_popneed['region'].isin(['ทั่วประเทศ'])].copy()

#Short labels
short_labels = {
    'แก้ไขปัญหาค่าครองชีพสูง': 'ค่าครองชีพสูง',
    'ปรับ/เพิ่มราคาพืชผลทางการเกษตร': 'พัฒนาเกษตร',
    'จัดหาอาชีพเสริมเพื่มรายได้': 'รายได้เสริม',
}
df_filtered['Short_Iss_Need'] = df_filtered['Iss_Need'].map(short_labels).fillna(df_filtered['Iss_Need'])

#Create bar chart
g = sns.catplot(
    data=df_filtered,
    kind='bar',
    x='Per_vill', y='Short_Iss_Need',
    hue='Need',
    col='region',
    sharex=True,
    height=4, aspect=1.2,
    palette='muted',
    legend_out=True,
    orient='h'
)

#Add label
for ax in g.fig.axes:
  for container in ax.containers:
     ax.bar_label(
          container,
         fmt='%.1f',
         padding=10,
          color='black',
         fontsize=8,
          weight='bold'
      )

#Adjust graph
g.set_axis_labels("ร้อยละของประชาชน", "")
g.set_titles("ร้อยละความต้องการของประชาชนที่ต้องการให้ภาครัฐเข้ามาช่วยเหลือ (ปี2566) ทั่วประเทศ", fontsize=18)
g.despine(left=True)
g._legend.set_bbox_to_anchor((1.25, 0.8))
g._legend.set_frame_on(True)
plt.tight_layout()


plt.savefig("G3.png", dpi=600, bbox_inches='tight')
plt.show()

#Save graph
#files.download("G3.png")
```

```python
#Filter data for analysis
df_filtered = df_popneed[df_popneed['region'].isin(['ยะลา', 'นราธิวาส', 'ปัตตานี','กรุงเทพมหานคร'])].copy()

#Short labels
short_labels = {
    'ปรับ/เพิ่มราคาพืชผลทางการเกษตรราคาตกต่ำ (ข้าว ยางพารา อ้อย ปาล์มน้ำมัน สัตว์น้ำ สัตว์บก ฯลฯ)': 'พัฒนาเกษตร',
    'แก้ไขปัญหาว่างงาน/ตกงาน': 'ปัญหาว่างงาน',
    'จัดหาอาชีพเสริมเพิ่มรายได้': 'รายได้เสริม',
    'แก้ไขปัญหาค่าครองชีพสูง': 'ค่าครองชีพสูง',
}
df_filtered['Short_Iss_Need'] = df_filtered['Iss_Need'].map(short_labels).fillna(df_filtered['Iss_Need'])

#Create bar chart
g = sns.catplot(
    data=df_filtered,
    kind='bar',
    x='Per_vill', y='Short_Iss_Need',
    hue='Need',
    col='region',
    sharex=True,
    height=4, aspect=0.7,
    palette='muted',
    legend_out=True,
    orient='h'
)

#Add label
for ax in g.fig.axes:
  for container in ax.containers:
     ax.bar_label(
          container,
         fmt='%.1f',
         padding=10,
          color='black',
         fontsize=8,
          weight='bold'
      )
     for spine in ax.spines.values():
        spine.set_visible(True)
        spine.set_linewidth(1.0)
        spine.set_edgecolor('black')


#Adjust graph
g.set_axis_labels("ร้อยละของประชาชน", "")
g.set_titles("{col_name}", size = "12")
g.despine(left=True)
g._legend.set_bbox_to_anchor((1.13, 0.8))
g._legend.set_frame_on(True)
plt.tight_layout()
plt.suptitle("ร้อยละความต้องการของประชาชนที่ต้องการให้ภาครัฐเข้ามาช่วยเหลือ (ปี2566)",size="18", y=1.14)
g.fig.subplots_adjust(wspace=0.2)

plt.savefig("G4.png", dpi=600, bbox_inches='tight')
plt.show()

#Save graph
#files.download("G4.png")
```

**กราฟ 5:** อัตราการย้ายถิ่นออกจากถิ่นเดิม จำแนกตามอันดับภูมิภาค ในปี 2565-2567

```python
#Filter data for analysis
df_move_from = pd.read_csv(io.BytesIO(move_from['ย้ายถิ่นมาจาก รายภาค.csv']))
df_source = df_move_from[df_move_from['typ_mig'].str.contains('ย้ายมาจาก')]
df_grouped = df_source.groupby(['year', 'typ_mig'])['value'].sum().reset_index()
df_grouped['rank'] = df_grouped.groupby('year')['value'].rank(ascending=False, method='first')
df_grouped['Rank_Only'] = 'อันดับ ' + df_grouped['rank'].astype(int).astype(str)
df_sorted = df_grouped.sort_values(['year', 'value'])

#Add label
def draw_barplot_with_labels(data, **kwargs):
    ax = sns.barplot(
        data=data,
        y='typ_mig',
        x='value',
        hue='Rank_Only',
        dodge=False,
        palette="muted"
    )
    for container in ax.containers:
        ax.bar_label(container, fmt='%.1f', label_type='edge', padding=3)

#Create bar chart
g = sns.FacetGrid(
    df_sorted,
    col="year",
    sharey=True,
    height=6,
    aspect=1
)
g.map_dataframe(draw_barplot_with_labels)

#Adjust graph
g.add_legend(title="อันดับ")
g._legend.set_bbox_to_anchor((1.0, 0.8))
g._legend.set_frame_on(True)
g._legend.get_frame().set_edgecolor('black')
g._legend.get_frame().set_linewidth(0.8)
g.set_axis_labels("จำนวน (พันคน)", "", size=12)
g.set_titles("ปี {col_name}", size=12)
plt.suptitle("อัตราการย้ายถิ่นออกจากถิ่นเดิม จำแนกตามอันดับภูมิภาค ในปี 2565-2567", size=18, y=1.03)

plt.tight_layout()
plt.savefig("G5.png", dpi=600, bbox_inches='tight')
plt.show()

#Save graph
#files.download("G5.png")
```

**กราฟ 6:** การย้ายถิ่นตามรูปแบบ แยกตามปีและรูปแบบการย้ายถิ่น

```python
#Filter data for analysis
df_move_household_filtered = df_move_household[df_move_household['year'].isin([2565, 2566, 2567])]
#Group by year and move_fo
grouped_data = df_move_household_filtered.groupby(['year', 'move_fo'])['value'].sum().unstack()

#Create bar chart
ax = grouped_data.plot(kind='bar', figsize=(10, 5.5),color=sns.color_palette("muted"), width=0.5)
ax.set_xlim()

#Adjust graph
plt.xlabel('ปี', fontproperties=font_prop,size = 11)
plt.ylabel('จำนวน (พันคน)', fontproperties=font_prop,size = 11)
plt.title('การย้ายถิ่นตามรูปแบบ แยกตามปีและรูปแบบการย้ายถิ่น', fontproperties=font_prop,size = 15)
plt.xticks(rotation=0, fontproperties=font_prop)
plt.tight_layout()
plt.legend(title='รูปแบบการย้ายถิ่น', prop=font_prop, bbox_to_anchor=(1.05, 1), loc='upper left')

#Add label
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height, f'{height:.1f}', ha='center', va='bottom', fontproperties=font_prop)

plt.savefig("G6.png", dpi=600, bbox_inches='tight')
plt.show()

#Save graph
#files.download("G6.png")
```

**กราฟ 7:** การย้ายถิ่นตามรูปแบบ แยกตามปีและรูปแบบการย้ายถิ่น

```python
#Filter data for analysis
df_filtered = df_move_reason[df_move_reason['year'].isin([2565, 2566, 2567])]
df_grouped = df_filtered.groupby(['year', 'rm_mun1'])['value'].sum().reset_index()
top5_each_year = (
    df_grouped
    .sort_values(['year', 'value'], ascending=[True, False])
    .groupby('year', group_keys=False)
    .head(5)
)

top5_each_year['rm_mun1'] = top5_each_year.groupby('year')['rm_mun1']\
    .transform(lambda x: pd.Categorical(x, categories=x, ordered=True))

#Create bar chart
g = sns.catplot(
    data=top5_each_year,
    kind='bar',
    x='rm_mun1',
    y='value',
    hue='rm_mun1',
    col='year',
    col_wrap=3,
    sharey=True,
    height=7,
    aspect=0.8,
    palette='muted',
    legend=False
)
#Add label
for ax in g.axes.flat:
    for bar in ax.patches:
        height = bar.get_height()
        if height > 0:
            ax.text(
                bar.get_x() + bar.get_width()/2, height + 5,
                f'{height:.1f}', ha='center', va='bottom', fontsize=10
            )
#Adjust graph
g.set_titles("ปี {col_name}",size="14")
g.set_axis_labels("", "จำนวน (พันคน)",size="12")
for ax in g.axes.flatten():
    ax.tick_params(axis='x', rotation=45)
plt.suptitle("สาเหตุการย้ายถิ่นทั่วประเทศ 5 อันดับแรก (ปี 2565-2567) ",size="18", y=1.0)

plt.tight_layout()
plt.savefig("G7.png", dpi=600, bbox_inches='tight')
plt.show()

#Save graph
#files.download("G7.png")
```
